## Exploring Price by Brand

Using aggregation, we explore the brand column to reveal unique values to aggregate by, create a storage for the value and calculate the mean of those unique values.

```python
#First off, we explore the "brand" column.
print(autos["brand"].describe())
brands = autos["brand"].value_counts().head(20)
```

The first twenty (20) cars are to be aggregated against ranging from the top brand i.e volkswagen to the 20th brand i.e mini.

```python
#We select the 20 most common cars as the dataset to work on aggregation with.
most_common_20_cars = brands.head(20).index
print(most_common_20_cars)
```

```python
brand_mean_prices = {}

for brand in most_common_20_cars:
     brand_row = autos[autos["brand"] == brand]
     mean_value = brand_row["price"].mean()
     brand_mean_prices[brand] = mean_value
    
print(brand_mean_prices)
    
```

From the above mean prices, lets work with the to six (6) brands:
1. The "audi", "BMW" and "Mercedes Benz" brands are typically high priced vehicles.
2. Budget friendly brands include "Ford", "Opel" etc.
3. The "Volkswagen" model exists in between.

## Storing Aggregate Data in a DataFrame

From the mean price analysis, we discovered the distinct price grap in the top 6 vehicle brands. We can use aggregation to understand the average mileage for those cars and connect any possible links. i.e Does more expensive mean more mileage?



```python
#First off, we'll be converting the brand_mean_prices dictionary into a series using the pandas series constructor.
bmp_series = pd.Series(brand_mean_prices)
print(bmp_series)

#Then using the pandas Dataframe constructor, we create a single column dataframe with a column title specified.
bmp_df = pd.DataFrame(bmp_series, columns=["mean_price"])
print(bmp_df)
```

```python
mean_mileage_dictionary = {}
mean_price_dictionary = {}


#Calculating the mean mileagenand price for the top 6 brands and storing in a dictionary.
for brand in most_common_20_cars[:6]:
     brand_row = autos[autos["brand"] == brand]
     mean_value = brand_row["odometer_km"].mean()
     mean_mileage_dictionary[brand] = mean_value
    
for brand in most_common_20_cars[:6]:
     brand_row = autos[autos["brand"] == brand]
     mean_value = brand_row["price"].mean()
     mean_price_dictionary[brand] = mean_value
        
print(mean_mileage_dictionary)
print(mean_price_dictionary)
```

```python
#Converting both dictionaries to a series constructor using the aforementioned pandas series constructor.
mmd_series = pd.Series(mean_mileage_dictionary)
print(mmd_series)
mpd_series = pd.Series(mean_price_dictionary)
print(mpd_series)
```

```python
#Creating a dataframe from the mean_mileage series object using the pandas dataframe constructor.
mmd_df = pd.DataFrame(mmd_series, columns=["mean_mileage"])
mmd_df["mean_price"] = mpd_series
print(mmd_df)
```
